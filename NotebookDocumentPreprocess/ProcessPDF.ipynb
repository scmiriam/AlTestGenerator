```python
import pytesseract
from pdf2image import convert_from_path

PDF_PATH = "/content/drive/MyDrive/TFG/Automated Testing in Microsoft Dynamics 365 Busines Central 1.pdf"
OUTPUT_MD = "output.md"


def is_separator(line):
    line = line.strip()
    return len(line) > 10 and all(c == "-" for c in line)


def clean_lines(text):
    return [l.strip() for l in text.split("\n") if l.strip()]


def process_pdf(pdf_path):
    pages = convert_from_path(pdf_path, dpi=150)  # más rápido que 300 DPI

    md_output = []

    for i, page in enumerate(pages):
        print(f"Procesando página {i+1}...")

        text = pytesseract.image_to_string(page)

        lines = clean_lines(text)

        if not lines:
            continue

        # 🟢 1. Primera línea = título
        title = lines[0]

        # 🟢 2. Detectar contenido (ignorando separador)
        content_lines = []
        for line in lines[1:]:
            if is_separator(line):
                continue
            content_lines.append(line)

        # 🟢 3. Convertir a Markdown
        md_output.append(f"## {title}\n")

        for line in content_lines:
            md_output.append(line)

        md_output.append("\n")

    return "\n".join(md_output)


if __name__ == "__main__":
    md = process_pdf(PDF_PATH)

    with open(OUTPUT_MD, "w", encoding="utf-8") as f:
        f.write(md)

    print("✅ Markdown generado correctamente")
```

```python
def reformat_markdown(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    processed_lines = []
    start_processing = False
    trash_patterns = [r"654043353F", r"paulagni99e@gnail,con", r"paulagn1996@gnail\.con"]

    for line in lines:
        clean_line = line.strip()

        # 1. Detectar el inicio: ## 1
        if clean_line == "## 1":
            start_processing = True

        if not start_processing:
            continue

        if clean_line.startswith("# @"):
            clean_line = clean_line.replace("# ", "")
            processed_lines.append(clean_line)
            continue

        # 2. Eliminar líneas de basura
        if any(re.search(p, clean_line) for p in trash_patterns):
            continue

        # 3. PROCESAR HEADERS (Aquí estaba el error)
        if clean_line.startswith("## "):
            content = clean_line.replace("## ", "").strip()

            # Regla: Solo número o solo texto (H1)
            is_only_digit = re.match(r"^\d+$", content)

            if is_only_digit:
                processed_lines.append(f"# {content}")
            else:
                # Mover número al final como Page
                page_match = re.search(r"(\d+)", content)
                if page_match:
                    page_number = page_match.group(1)
                    clean_text = content.replace(page_number, "").strip(" ._—~-")
                    if not clean_text:
                        processed_lines.append(f"# {page_number}")
                    else:
                        processed_lines.append(f"## {clean_text} - Page: {page_number}")
                else:
                    processed_lines.append(f"## {content}")

        else:
              processed_lines.append(line.rstrip())


    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("\n".join(processed_lines))

def clip_to_main_content(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    final_lines = []
    keep = False
    for line in lines:
        clean_line = line.strip()
        if clean_line == "# 1":
            keep = True
        if clean_line == "## Appendix":
            break
        if keep:
            final_lines.append(line)

    with open(output_file, 'w', encoding='utf-8') as f:
        f.writelines(final_lines)

def fix_repeated_headers(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    chapter_titles = {}

    # --- PASO A: Escanear para encontrar los títulos de cada # ---
    current_chapter_id = None
    for i in range(len(lines)):
        line_s = lines[i].strip()
        # Detectar el marcador de capítulo # 1, # 2...
        chapter_match = re.match(r"^# (\d+)$", line_s)

        if chapter_match:
            current_chapter_id = line_s
            # Buscamos el PRIMER ## que aparezca después, sin límite de 5 líneas
            for j in range(i + 1, len(lines)):
                candidate = lines[j].strip()
                if candidate.startswith("# ") and candidate != current_chapter_id:
                    break # Si llegamos al siguiente #, paramos
                if candidate.startswith("## "):
                    # Extraer el título antes del " - Page:"
                    title_part = candidate.replace("## ", "").split(" - Page:")[0].strip()
                    chapter_titles[current_chapter_id] = title_part
                    break

    # --- PASO B: Reestructurar el archivo ---
    new_lines = []
    current_chapter_num = None

    for i, line in enumerate(lines):
        line_s = line.strip()

        # 1. Si es un H1 (# 1), le ponemos su título
        if re.match(r"^# (\d+)$", line_s):
            current_chapter_num = line_s
            full_title = f"{line_s} {chapter_titles.get(line_s, '')}".strip()
            new_lines.append(full_title)
            continue

        # 2. Si es un H2 (##), vemos si es el repetido
        if line_s.startswith("## "):
            page_part = ""
            if " - Page:" in line_s:
                content_part, page_part = line_s.split(" - Page:")
                page_part = " - Page:" + page_part
            else:
                content_part = line_s

            clean_content = content_part.replace("## ", "").strip()
            chapter_main_title = chapter_titles.get(current_chapter_num, "---")

            # Si el header es igual al título del capítulo (es el repetido de página par)
            if clean_content == chapter_main_title:
                found_better = False
                # Buscamos el siguiente ## real para "robarle" el nombre
                for k in range(i + 1, len(lines)):
                    next_l = lines[k].strip()
                    if next_l.startswith("# "): break # No saltar al siguiente capítulo
                    if next_l.startswith("## "):
                        cand_title = next_l.replace("## ", "").split(" - Page:")[0].strip()
                        if cand_title != chapter_main_title and "Summary" not in cand_title:
                            new_lines.append(f"## {cand_title}{page_part}")
                            found_better = True
                            break

                if not found_better:
                    new_lines.append(line_s)
            else:
                new_lines.append(line_s)

        # 3. SI ES CONTENIDO (párrafos, código, etc.), LO MANTENEMOS
        else:
            new_lines.append(line.rstrip())

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("\n".join(new_lines))

def fix_test_examples_and_pages(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    fixed_lines = []
    last_valid_page = 0
    for line in lines:
        line = line.strip()
        if not line: continue
        if line.startswith("# "):
            fixed_lines.append(line)
            continue

        if line.startswith("## "):
            page_match = re.search(r"Page:\s*(\d+)", line)
            proposed_page = int(page_match.group(1)) if page_match else None
            text_content = re.sub(r"##\s*", "", line)
            text_content = re.sub(r" - Page:.*", "", text_content)
            text_content = text_content.strip("'‘")
            text_content = text_content.replace("~", "-")
            text_content = re.sub(r"\s+\d+$", "", text_content)

            if proposed_page is not None:
                if proposed_page <= last_valid_page or (proposed_page - last_valid_page) > 10:
                    actual_page = last_valid_page + 1 if (last_valid_page > 0) else proposed_page
                else:
                    actual_page = proposed_page
            else:
                actual_page = last_valid_page + 1
            last_valid_page = actual_page
            fixed_lines.append(f"## {text_content} - Page: {actual_page}")
        else:
            fixed_lines.append(line)

    with open(output_file, 'w', encoding='utf-8') as f:
        f.write("\n".join(fixed_lines))
    return output_file

def unify_pdf_processing(markdown, final_output="resultado_final.md"):
    """
    Unifies the entire PDF processing workflow from OCR to final Markdown content.
    Returns: Final processed Markdown content.
    """
    temp_files = {
        "raw_ocr": "output.md",
        "reformatted": "salida_reestructurada.md",
        "clipped": "eliminadocontenido.md",
        "fixed_headers": "archivo_final_limpio.md",
        "fixed_pages": "paginas.md"
    }


    print("--- Step 1: Reformatting ---")
    reformat_markdown(markdown, temp_files["reformatted"])

    print("--- Step 2: Clipping ---")
    clip_to_main_content(temp_files["reformatted"], temp_files["clipped"])

    print("--- Step 3: Fixing Repeated Headers ---")
    fix_repeated_headers(temp_files["clipped"], temp_files["fixed_headers"])

    print("--- Step 4: Fixing Pages and Examples ---")
    final_path = fix_test_examples_and_pages(temp_files["fixed_headers"], final_output)

    # Optional: Clean up temporary files
    for f in temp_files.values():
         if os.path.exists(f):
            os.remove(f)

    print("Done!")
    print(f"--- Process Complete! Output saved to: {final_path} ---")
    return final_path

if __name__ == "__main__":
    markdown = "output.md"
    Result = unify_pdf_processing(markdown)
```
